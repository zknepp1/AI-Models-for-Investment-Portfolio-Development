```python

pattern = r'\d\d\d\d-\d\d-\d\d'

re_matches = []

for value in df['dt']:
  value = str(value)
  match = re.search(pattern, value)
  if match:
    re_matches.append(match.group())
  else:
    re_matches.append('no match')


df['clean_dates'] = re_matches
```

```python
cleaned_df = df.dropna()
```

```python
# Group by 'date_column' and aggregate text using join
grouped = cleaned_df.groupby('clean_dates').agg({'headline': ' '.join, 'snippet': ' '.join})


grouped
```

```python
grouped.to_csv('/content/drive/My Drive/news/groupeddata.csv')
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer

head_pos = []
head_neg = []
head_neu = []

snippet_pos = []
snippet_neg = []
snippet_neu = []


for index, row in grouped.iterrows():
  text1 = row['headline']
  text2 = row['snippet']

  # Create a SentimentIntensityAnalyzer object
  sid = SentimentIntensityAnalyzer()

  # Perform sentiment analysis
  sentiment_scores = sid.polarity_scores(text1)
  head_pos.append(sentiment_scores['pos'])
  head_neg.append(sentiment_scores['neg'])
  head_neu.append(sentiment_scores['neu'])

  sentiment_scores = sid.polarity_scores(text2)
  snippet_pos.append(sentiment_scores['pos'])
  snippet_neg.append(sentiment_scores['neg'])
  snippet_neu.append(sentiment_scores['neu'])



# Print the sentiment and scores
#print("Sentiment:", sentiment)
#print("Positive Score:", sentiment_scores['pos'])
#print("Negative Score:", sentiment_scores['neg'])
#print("Neutral Score:", sentiment_scores['neu'])
#print("Compound Score:", sentiment_scores['compound'])
```

```python
grouped['head_pos'] = head_pos
grouped['head_neg'] = head_neg
grouped['head_neu'] = head_neu

grouped['snippet_pos'] = snippet_pos
grouped['snippet_neg'] = snippet_neg
grouped['snippet_neu'] = snippet_neu
```
